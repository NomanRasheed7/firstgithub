# Matplotlib subplot
## Display Multiple plots

```python
import matplotlib.pyplot as plt
import numpy as np

#plot : 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("plot 1,color="blue")
plt.show()


#plot : 2
x = np.array([0,1,2,3])
y = np.array([3,4,6,1])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()
```

# Matplotlib Scatter

```python
#Creating scatter plots
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.title("Scatter plto",color="blue",size='20')
plt.xlabel("xaxis",color ="blue",size='20')
plt.ylabel("yaxis",color ="blue",size='20')
plt.show()
```

# Matplotlib Bars

## creating bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A","B","C","D"])
y = np.array([3,5,8,10])
colors = np.array(["red","green","blue"])

plt.bar(x,y,color="red")
plt.title("Creating Bar",color="blue",size='20')
plt.xlabel("x axis",color="Blue",size='20')
plt.ylabel("y axis",color="green",size='20')
plt.show()
```
